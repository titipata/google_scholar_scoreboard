```python
# import library to scrape website
from urllib2 import urlopen
import numpy as np
import pandas as pd
import time
from lxml import etree # read html
from lxml import html

# import library to plot and draw
import matplotlib.pyplot as plt
import cStringIO # get image from website
import Image

# use pygame to play music
import pygame


#import twitter
#api = twitter.Api(consumer_key='',
#                    consumer_secret='',
#                    access_token_key='',
#                    access_token_secret='')

#### CONSTANT ####
# name list
NAME = ['Konrad', 
        'Daniel', 
        'Pavan',
        'Josh', 
        'Ted', 
        'Pat',
        'Eva', 
        'Mohammad', 
        'Hugo',
        'Luca',
        'Sohrob',
        'Steve',
        'Iris']

# url of each lab member (based on name lists)
BASE_URL = ['http://scholar.google.com/citations?user=MiFqJGcAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=GAi23ssAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=JtltLUAAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=tbfWCDgAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=T8W-5LsAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=jjvixpcAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=wdFV87UAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=AlTQrFcAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=JG7xb2AAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=xxDk3-EAAAAJ&hl=en',
            'http://scholar.google.co.uk/citations?user=9jqURCEAAAAJ&hl=en',
            'http://scholar.google.com/citations?user=uwpOnSAAAAAJ&hl=en&oi=sra',
            'http://scholar.google.com/citations?user=Ztwn608AAAAJ&hl=en']


# image link that we want to display if someone get cited
IMG_LINK = {'Konrad': 'http://www.qwantz.com/patreon/p3.png',
            'Daniel': 'http://www.quickmeme.com/img/45/451ab8e56df6f66c37c7eda8e36765f743cbabf1e1dbee5dffd648f47dde54d1.jpg',
            'Mohammad': 'http://mybroadband.co.za/vb/attachment.php?s=c864c202183cf1b3d2c57f738e78fce8&attachmentid=103452&d=1394014893',
            'Pavan': 'http://www.nbc.com/sites/nbcunbc/files/files/styles/nbc_bio_image/public/images/2013/11/08/azizAnsari_tomHaverford.jpg?itok=PCowY6uk',
            'Hugo': 'http://public.media.smithsonianmag.com/legacy_blog/dinosaur-comic-strip.jpg',
            'Ted': 'http://lovestats.files.wordpress.com/2012/07/r-square-success-kid.jpg',
            'Pat': 'http://m.memegen.com/x6259d.jpg',
            'Josh': 'http://cdn.mhpbooks.com/uploads/2013/10/Success-Kid.jpg',
            'Eva': 'http://veryhilarious.com/wp-content/uploads/2012/07/indy-hipster.jpg',
            'Luca': 'http://i1.cpcache.com/product_zoom/510200164/bunga_bunga_berlusconi_classic_thong.jpg?color=White&height=460&width=460&padToSquare=true',
            'Sohrob': 'http://4.bp.blogspot.com/-rjyBJpjUizw/U35ON10k8eI/AAAAAAAAfUM/KMFJEgwJhM8/s1600/stupid-meme-stalin-obama-2.jpg',
            'Steve': 'http://ct.fra.bz/ol/fz/sw/i58/2/5/25/frabz-giant-burrito-man-will-make-you-fire-torpedoes-of-another-kind-a01c69.jpg',
            'Iris': 'http://public.media.smithsonianmag.com/legacy_blog/dinosaur-comic-strip.jpg'
            }


# music snippet that you want to play if anyone got cited
MUSIC_DIR = '/Users/titipat/Desktop/Amazon Web Service/snippet.wav'

# parameter for plotting
params_cite = {'fontsize': 22, 'color': 'green', 'fontweight':'bold'}
params_hindex = {'fontsize': 22, 'color': 'red', 'fontweight':'bold'}
params_date = {'fontsize': 20, 'color': 'blue', 'fontweight':'bold'}
params_others = {'fontsize': 30, 'color': 'black'}
params_gini = {'fontsize': 20, 'color': 'blue', 'fontweight':'bold'}
params_gini_val = {'fontsize': 20, 'color': 'black'}
params_tweet = {'fontsize': 20, 'color': 'red', 'fontweight':'bold'}

def get_citation_matrix():
    ''' Get all citation datafram from Google Scholar '''
    all_people = pd.DataFrame(columns=['name', 'citation', 'h_index','url'])
    all_people['name'] = NAME
    all_people['url'] = BASE_URL

    for i in range(len(all_people)):
        tree = html.parse(all_people.url[i])
        cit = tree.xpath("/html/body/div[@id='gs_top']/div[@id='gsc_bdy']/div[@id='gsc_rsb']/div[@class='gsc_rsb_s']/table[@id='gsc_rsb_st']//tr//td[@class='gsc_rsb_std']")
        citations = np.int(cit[0].text)
        h_index = np.int(cit[2].text)
        
        all_people['citation'][i] = np.int(citations)
        all_people['h_index'][i] = np.int(h_index)

    return all_people # return matrix of citation

def sort_citation(all_people):
    ''' Sort the given dataframe by citation '''
    all_people_sorted = all_people.sort(columns=['h_index', 'citation'], ascending=False)
    all_people_sorted.index = np.arange(len(all_people))
    return all_people_sorted

def get_options(all_people, all_people_new):
    ''' Get citation and new citation then return option dataframe (including difference) '''
    # get difference
    cite_diff = all_people_new.citation - all_people.citation
    index = cite_diff.nonzero()[0]
    name_diff = list(all_people.name[index]) # list of name different
    
    # get table of options if there is different
    options = pd.DataFrame(columns=['name', 'citation_diff', 'h_index_diff','sign'])
    options['name'] = all_people['name'] # refer to new update
    options['citation_diff'] = 0
    options['h_index_diff'] = 0
    options['sign'] = ''

    # for one people in name_diff
    for j in range(len(name_diff)):
        idx_new = np.where(name_diff[j] == all_people_new['name'])[0]
        idx_old = np.where(name_diff[j] == all_people['name'])[0]
        diff_citation = all_people_new.citation[idx_new] - all_people.citation[idx_old]
        diff_hindex = all_people_new.h_index[idx_new] - all_people.h_index[idx_old]
        #print sign(all_people_sorted_new.citation[idx_new] - all_people_sorted.citation[idx_old]) # get sign +1 or -1
        options['citation_diff'][idx_new] = diff_citation
        options['h_index_diff'][idx_new] = diff_hindex
        sign_pm = np.int(np.sign(diff_citation))
        if sign_pm == 1:
            options['sign'][idx_new] = '+'
        else:
            options['sign'][idx_new] = '-'

    # change index
    options.index = options.name # do this one!
    
    # return dataframe of options and name that have different citation
    return options, name_diff
    
def is_different(all_people_sorted, all_people_sorted_new):
    ''' Find if two dataframe are equal or not '''
    result = np.max(all_people_sorted_new.h_index != all_people_sorted.h_index) or np.max(all_people_sorted_new.citation != all_people_sorted.citation)
    return result

def play_music():
    ''' Play Everything is Awesome wavfile track '''
    pygame.init()
    pygame.mixer.music.load(MUSIC_DIR)
    pygame.mixer.music.play()
    
def compute_gini(y):
    ''' function to compute Gini index '''
    N = len(y)
    gini = double(2*np.dot(sorted(y, reverse=False), np.arange(1, N+1)))/double(double(N)*np.sum(y)) - ((N+1.0)/double(N))
    gini = np.ceil(gini * 1000) / 1000.0
    return gini
    
def draw_citation_table(): #all_people_new_sorted
    global all_people, all_people_new, all_people_sorted, all_people_new_sorted, options, options_old, name_diff # see the outer variable
    
    # get new citation
    all_people_new = get_citation_matrix()
    all_people_new_sorted = sort_citation(all_people_new)
    
    # if different optain new options
    if is_different(all_people_sorted, all_people_new_sorted):
        options_old, name_diff = get_options(all_people, all_people_new)
        options = options_old
    else:
        options = options_old
     
    
    # DRAW TITLE/ HEADERS
    fig = plt.gcf() # get current figure
    fig.clf() # clear current figure
    fig.suptitle('Bayesian Behavior Lab Citations',
                 fontsize=35, fontweight='bold',
                 color='gray', style='italic')
    
    plt.text(0.2, 0.94, 'Citations', **params_cite)
    plt.text(0.5, 0.94, 'h-index', **params_hindex)
    plt.axis('off')

    for i in range(len(all_people_new_sorted)):
        name = all_people_new_sorted.name[i] # get name
        if (name in name_diff):
            option_cit = '('+ options.loc[name].sign + str(options.loc[name].citation_diff) + ')'
            option_h = '(' + options.loc[name].sign + str(options.loc[name].h_index_diff) + ')'
            # if citation or h-index different is 0, turn to blank
            if np.int(options.loc[name].citation_diff) == 0:
                option_cit = ''
            if np.int(options.loc[name].h_index_diff) == 0:
                option_h = ''
        else:
            option_cit = ''
            option_h = ''

        # DRAW Name, distance between shown name lists is here
        plt.text(-0.1, 0.85-0.065*i, str(all_people_new_sorted.name[i]) , **params_others)
        # DRAW Citation
        plt.text(0.2, 0.85-0.065*i, str(all_people_new_sorted.citation[i]) + option_cit, **params_others)
        # DRAW H-index
        plt.text(0.5, 0.85-0.065*i, str(all_people_new_sorted.h_index[i]) + option_h, **params_others)

    plt.text(0.0, -0.1, 'Last citation update: ' + time.strftime('%X %b %d, %Y'), **params_date)
    plt.text(0.65, 0.6, 'Gini (citation)', **params_gini)
    plt.text(0.92, 0.6, str(compute_gini(all_people_new_sorted.citation)), **params_gini_val)
    plt.text(0.65, 0.53, 'Gini (h-index)', **params_gini)
    plt.text(0.92, 0.53, str(compute_gini(all_people_new_sorted.h_index)), **params_gini_val)

    # DRAW New Twitter Feed from Kording Lab
    #statuses = api.GetUserTimeline(screen_name="KordingLab") # getting all tweets
    #plt.text(-0.1, -0.05, 'Tweets: ', **params_gini)
    #try:
    #    new_tweet = str(statuses[0].text.encode('ascii', 'ignore'))
    #    plt.text(0.1, -0.05, new_tweet, **params_gini_val)
    #except ValueError:
    #    plt.text(0.1, -0.05, "Can't retrive tweet...", **params_gini_val)
    
    
    # DRAW IMAGE
    if len(name_diff) > 0:
        URL = IMG_LINK[name_diff[-1]] # use lowested name to show image
    else:
        URL = 'http://www.qwantz.com/patreon/p3.png' # default dinosaur images 
    file = cStringIO.StringIO(urlopen(URL).read())

    img = Image.open(file)
    axicon = fig.add_axes([0.6,0.15,0.33,0.33])
    plt.imshow(img)
    plt.axis('off')
    
    fig.canvas.draw()
    fig.canvas.activateWindow()
    plt.draw()
    plt.show()
    
    
    # Play music and update part!
    if is_different(all_people_sorted, all_people_new_sorted):
        play_music()
        all_people = all_people_new # replace all people with new one
        all_people_sorted = sort_citation(all_people) # sorted again
```

# Run Google Scholar

After running library and functions part, run this line to show the citation update

```python
# Draw the First citation image
plt.close('all')
fig = plt.figure(facecolor='white') # or 'white' depending on bg we want

all_people = get_citation_matrix() # get citation from provided url
options, name_diff = get_options(all_people, all_people)
options_old = options # assign value to get rid of conflict
all_people_sorted = sort_citation(all_people)
all_people_new = get_citation_matrix()
all_people_new_sorted = sort_citation(all_people_new)
draw_citation_table() # draw first K-lab citation

# timer to run code every some amount of time
timer = fig.canvas.new_timer(interval=1000*60*15) # run every 15 minutes (1000*60*15 seconds)
timer.add_callback(draw_citation_table)
timer.start()
```
